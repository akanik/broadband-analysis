```python
import pandas as pd
import glob, os
import geopandas as gpd
```

```python
geocorr_dir = '../../../GIS/geocorr-blocks-places'
```

```python
df_list = []

for file in os.listdir(geocorr_dir):
    print(file)
    if file.endswith('.csv'):
        df = pd.read_csv(geocorr_dir+'/'+file)
        df_list.append(df)
        
all_corr = df = pd.concat(df_list)
all_corr.to_csv(geocorr_dir+'/geocorr2018-block-place-all.csv')
```
